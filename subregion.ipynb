### 1. Adding subregion to ads sheet

```python
import pandas as pd
import os
```

```python
# subregion dataframe
path = os.getcwd()
sr_df = pd.read_excel(f"{path}/data/country_codes.xlsx")
sr_df = sr_df[['ISO3166-1-Alpha-2', 'Sub-region Name']]
sr_df = sr_df.fillna("CQ")

sr_df
```

```python
ads_df = pd.read_csv(f"{path}/intermediate/ads_data.csv")

ads_df
```

```python
subregion = []

subregion_list = sr_df["Sub-region Name"].tolist()
iso_code = sr_df["ISO3166-1-Alpha-2"].tolist()

main_country = ads_df['delivery_country']

# map main_country to subregion
for i in range(len(main_country)):
    target = main_country[i]

    for j in range(len(iso_code)):
        code = iso_code[j]

        if code == target:
            subregion.append(subregion_list[j])

subregion[:10]
```

```python
ads_df["Subregion"] = subregion

ads_df
```

```python
ads_df.to_excel('intermediate/ads_data_final.xlsx')
```

### 2. Adding subregion to Moderators sheet

```python
mods_df = pd.read_csv(f"{path}/intermediate/moderators_data.csv")

mods_df
```

```python
mkt_list = mods_df['market']
mkt_list_string = mods_df['market']
subregion_list = sr_df["Sub-region Name"].tolist()
iso_code = sr_df["ISO3166-1-Alpha-2"].tolist()
subregion = []

# separate individual markets
for i in range(1414):
    mkt_string = mkt_list[i]
    mkt_string = mkt_string.strip("[] \"")
    mkt_string = mkt_string.split(',')

    clean_mkt_list = []

    for code in mkt_string:
        code = code.strip("[] \"")
        code = code.strip(" ")
        clean_mkt_list.append(code)
    
    subregion.append(clean_mkt_list)
```

```python
subregion_column = []

for j in range(1414):
    row_mktlist = subregion[j]
    sr_row = []
    
    # map each market to subregion
    for country in row_mktlist:
        for a in range(len(iso_code)):
            iso = iso_code[a]
            sr = subregion_list[a]

            if country == iso and sr not in sr_row:
                sr_row.append(sr)
                break
    
    subregion_column.append(sr_row)
```

```python
mods_df['Subregion'] = subregion_column
mods_df.info()
```

```python
ls = []

# get all subregions
for i in range(len(ads_df)):
    row = ads_df['Subregion'].iloc[i]
    if row not in ls:
        ls.append(row)

# create output excel sheet
excel_name = "intermediate/mods_data_final.xlsx"
blank_df = pd.DataFrame().to_excel(excel_name)

for region_name in ls:
    useful_df = pd.DataFrame()
    for i in range(1414):
        row = mods_df.iloc[i]
        subregion_internal_list = row['Subregion']

        for reg in subregion_internal_list:
            if region_name in reg:  
                row['Subregion'] = ", ".join(subregion_internal_list)
                row['market'] = mkt_list_string[i]
                useful_df = pd.concat([useful_df,pd.DataFrame(row).T])

    writer = pd.ExcelWriter(excel_name, engine = 'openpyxl', mode='a')
    output_df = useful_df.drop_duplicates()
    output_df.to_excel(writer, sheet_name=region_name)
    writer.close()
```
